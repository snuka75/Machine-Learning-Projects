```python
import zipfile
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

# Step 1: Extract the zip file
def extract_zip(zip_file_path, extract_dir):
    # Check if the extraction directory exists; if not, create it
    if not os.path.exists(extract_dir):
        os.makedirs(extract_dir)

    # Extract the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)

    # List the extracted files to verify
    extracted_files = os.listdir(extract_dir)
    return extracted_files

# Step 2: Load and preprocess data
def load_and_preprocess_data(filenames):
    data = {}
    for file in filenames:
        df = pd.read_csv(file, usecols=['Date', 'Close'])
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        data[file] = df
    return data

# Prepare the data for LSTM model (scaling and reshaping)
def prepare_data_for_lstm(data, window_size=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)

    X, y = [], []
    for i in range(window_size, len(scaled_data)):
        X.append(scaled_data[i-window_size:i, 0])
        y.append(scaled_data[i, 0])

    X = np.array(X)
    y = np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    return X, y, scaler

# Step 3: Build and train the LSTM model
def build_and_train_lstm(X, y):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=25))
    model.add(Dense(units=1))

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(X, y, batch_size=32, epochs=10)
    return model

# Step 4: Predict and visualize with accuracy metrics
def plot_predictions_with_accuracy(model, X, y, scaler, data, company_name):
    predicted_stock_price = model.predict(X)
    predicted_stock_price = scaler.inverse_transform(predicted_stock_price)

    # Adjust the data for plotting
    actual_stock_price = data[len(data) - len(predicted_stock_price):]  # Slice the actual data to match the length of predictions

    # Calculate accuracy metrics
    mae = mean_absolute_error(actual_stock_price, predicted_stock_price)
    mse = mean_squared_error(actual_stock_price, predicted_stock_price)
    rmse = math.sqrt(mse)

    # Plot the results
    plt.plot(actual_stock_price, label='Actual Stock Price')
    plt.plot(predicted_stock_price, label='Predicted Stock Price')
    plt.title(f'Stock Price Prediction for {company_name}')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

    # Print accuracy metrics
    print(f"Accuracy for {company_name}:")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print("-" * 50)

# Main function to execute the steps
def main():
    # Zip file and extraction directory
    zip_file_path = '/content/archive.zip'  # Replace with the actual path to your zip file
    extract_dir = '/content/archive/'    # Replace with the actual extraction directory

    # Step 1: Extract zip file
    extracted_files = extract_zip(zip_file_path, extract_dir)
    print("Extracted files:", extracted_files)

    # Step 2: Filter the files to sectors mentioned (Auto, Banking, Media, Energy)
    auto_sector = ['BAJAJ-AUTO.csv', 'HEROMOTOCO.csv', 'TATAMOTORS.csv', 'EICHERMOT.csv', 'MARUTI.csv']
    banking_sector = ['AXISBANK.csv', 'HDFCBANK.csv', 'ICICIBANK.csv', 'INDUSINDBK.csv', 'KOTAKBANK.csv', 'SBIN.csv']
    media_sector = ['ZEEL.csv']
    energy_sector = ['BPCL.csv', 'GAIL.csv', 'ONGC.csv', 'POWERGRID.csv', 'RELIANCE.csv']

    # Combine all sectors into one list
    all_sectors = auto_sector + banking_sector + media_sector + energy_sector
    data_dir = extract_dir
    files = [data_dir + filename for filename in all_sectors if filename in extracted_files]

    # Step 3: Load and preprocess data
    data = load_and_preprocess_data(files)

    # Debug print to check available data keys
    print("Available data keys:", data.keys())

    # Step 4: Iterate through each sector and generate predictions and plots
    for sector in data.keys():
        print(f"Processing {sector}...")
        stock_data = data[sector]['Close'].values.reshape(-1, 1)
        X, y, scaler = prepare_data_for_lstm(stock_data)

        # Build and train the LSTM model
        model = build_and_train_lstm(X, y)

        # Predict and visualize with accuracy metrics
        plot_predictions_with_accuracy(model, X, y, scaler, stock_data, sector)

# Run the main function
if __name__ == "__main__":
    main()

```

```python
import zipfile
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

# Step 1: Extract the zip file
def extract_zip(zip_file_path, extract_dir):
    # Check if the extraction directory exists; if not, create it
    if not os.path.exists(extract_dir):
        os.makedirs(extract_dir)

    # Extract the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)

    # List the extracted files to verify
    extracted_files = os.listdir(extract_dir)
    return extracted_files

# Step 2: Load and preprocess data
def load_and_preprocess_data(filenames):
    data = {}
    for file in filenames:
        df = pd.read_csv(file, usecols=['Date', 'Close'])
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        data[file] = df
    return data

# Prepare the data for LSTM model (scaling, splitting, and reshaping)
def prepare_data_for_lstm(data, window_size=60, split_ratio=0.8):
    # Scale the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)

    # Split data into training and testing sets
    training_data_len = int(np.ceil(len(scaled_data) * split_ratio))
    train_data = scaled_data[:training_data_len, :]
    test_data = scaled_data[training_data_len - window_size:, :]  # Include overlap for test set

    # Prepare training data
    X_train, y_train = [], []
    for i in range(window_size, len(train_data)):
        X_train.append(train_data[i-window_size:i, 0])
        y_train.append(train_data[i, 0])

    X_train, y_train = np.array(X_train), np.array(y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

    # Prepare testing data
    X_test, y_test = [], []
    for i in range(window_size, len(test_data)):
        X_test.append(test_data[i-window_size:i, 0])
        y_test.append(test_data[i, 0])

    X_test, y_test = np.array(X_test), np.array(y_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    return X_train, y_train, X_test, y_test, scaler

# Step 3: Build and train the LSTM model
def build_and_train_lstm(X, y):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=25))
    model.add(Dense(units=1))

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(X, y, batch_size=32, epochs=35)
    return model

# Step 4: Calculate accuracy metrics and plot predictions
def plot_predictions_with_accuracy(model, X_train, X_test, y_train, y_test, scaler, data, company_name):
    # Predict on both train and test data
    predicted_train = model.predict(X_train)
    predicted_train = scaler.inverse_transform(predicted_train)

    predicted_test = model.predict(X_test)
    predicted_test = scaler.inverse_transform(predicted_test)

    # Actual test stock prices
    actual_test = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Calculate accuracy metrics
    mae = mean_absolute_error(actual_test, predicted_test)
    mse = mean_squared_error(actual_test, predicted_test)
    rmse = math.sqrt(mse)

    # Calculate MAPE and Accuracy
    mape = np.mean(np.abs((actual_test - predicted_test) / actual_test)) * 100
    accuracy = 100 - mape

    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(data, label='Actual Stock Price')
    plt.plot(np.arange(len(data) - len(actual_test), len(data)), actual_test, label='Test Actual')
    plt.plot(np.arange(len(data) - len(actual_test), len(data)), predicted_test, label='Test Predicted')
    plt.title(f'Stock Price Prediction for {company_name}')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

    # Print accuracy metrics for test data
    print(f"Test Accuracy for {company_name}:")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape}%")
    print(f"Accuracy: {accuracy}%")
    print("-" * 50)

# Main function to execute the steps
def main():
    # Zip file and extraction directory
    zip_file_path = '/content/archive.zip'  # Replace with the actual path to your zip file
    extract_dir = '/content/archive/'    # Replace with the actual extraction directory

    # Step 1: Extract zip file
    extracted_files = extract_zip(zip_file_path, extract_dir)
    print("Extracted files:", extracted_files)

    # Step 2: Filter the files to sectors mentioned (Auto, Banking, Media, Energy)
    auto_sector = ['BAJAJ-AUTO.csv', 'HEROMOTOCO.csv', 'TATAMOTORS.csv', 'EICHERMOT.csv', 'MARUTI.csv']
    banking_sector = ['AXISBANK.csv', 'HDFCBANK.csv', 'ICICIBANK.csv', 'INDUSINDBK.csv', 'KOTAKBANK.csv', 'SBIN.csv']
    media_sector = ['ZEEL.csv']
    energy_sector = ['BPCL.csv', 'GAIL.csv', 'ONGC.csv', 'POWERGRID.csv', 'RELIANCE.csv']

    # Combine all sectors into one list
    all_sectors = auto_sector + banking_sector + media_sector + energy_sector
    data_dir = extract_dir
    files = [data_dir + filename for filename in all_sectors if filename in extracted_files]

    # Step 3: Load and preprocess data
    data = load_and_preprocess_data(files)

    # Debug print to check available data keys
    print("Available data keys:", data.keys())

    # Step 4: Iterate through each sector and generate predictions and plots
    for sector in data.keys():
        print(f"Processing {sector}...")
        stock_data = data[sector]['Close'].values.reshape(-1, 1)
        X_train, y_train, X_test, y_test, scaler = prepare_data_for_lstm(stock_data)

        # Build and train the LSTM model
        model = build_and_train_lstm(X_train, y_train)

        # Predict and visualize with accuracy metrics
        plot_predictions_with_accuracy(model, X_train, X_test, y_train, y_test, scaler, stock_data, sector)

# Run the main function
if __name__ == "__main__":
    main()
```
